## **Create a bar graph comparing highest and lowest prices to the manufacturers column of the dataframe.**

```python
# read in libraries

import pandas as pd
import numpy as np
from pathlib import Path
import logging
import matplotlib as plt
import matplotlib.pyplot as plt
import plotly.express as px

# read in dataframe
# drop title column
# convert price to int32
# sort the values in ascending order

price_df = pd.read_csv("guitars.csv")
price_df = price_df.drop('title', 1)
price_df['price'] = price_df['price'].str.replace(',', '').astype(int)
price_df = price_df.sort_values(['price'], ascending=[False])


print(price_df.head(10))
```

```python
# plot a bar graph of highest prices

price_df.head(15).plot.bar(title = "Highest Prices", x = "manufacturer", y = "price", xlabel = "Manufacturers", ylabel = "Prices", rot = 'vertical')


```

```python
# print the lowest price rows

print(price_df.tail(15))
```

```python
# plot a bar graph of highest prices


price_df.tail(15).plot.bar(title = "Lowest Prices", x = "manufacturer", y = "price", xlabel = "Manufacturers", ylabel = "Prices", rot = 'vertical')
```

## *In the dataframe analyzed we see that the highest priced guitars are made by Gibson and the lowest priceed guitars are made by Harley Benton.

```python
#  #what is the average cost of guitars in this dataframe

price_df.mean().plot.bar(title = 'Average', x = ' ', y = 'price', xlabel = 'manufacturer', ylabel = 'price')
```
